```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image = mpimg.imread('test.jpg')

print('This image is: ', type(image), 'with dimentions:', image.shape)

ysize = image.shape[0]
xsize = image.shape[1]

color_select = np.copy(image)
```

```python
red_threshold = 200
green_threshold = 200
blue_threshold = 200

rgb_threshold = [red_threshold, green_threshold, blue_threshold] 
```

```python
thresholds = (image[:,:,0] < rgb_threshold[0]) \
            | (image[:,:,1] < rgb_threshold[1]) \
            | (image[:,:,2] < rgb_threshold[2])
        
color_select[thresholds] = [0,0,0]

plt.imshow(image)
plt.show()

plt.imshow(color_select)
plt.show()
```

```python
left_bottom = [100, 539]
right_bottom = [900, 539]
apex = [450, 320]

fit_left = np.polyfit((left_bottom[0], apex[0]), (left_bottom[1], apex[1]), 1)
fit_right = np.polyfit((right_bottom[0], apex[0]), (right_bottom[1], apex[1]), 1)
fit_bottom = np.polyfit((left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1)

XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))

region_thresholds = (YY > (XX * fit_left[0] + fit_left[1])) & \
                    (YY > (XX * fit_right[0] + fit_right[1])) & \
                    (YY < (XX * fit_bottom[0] + fit_bottom[1]))
        
print(region_thresholds)
```

```python
region_select = np.copy(image)

region_select[region_thresholds] = region_select[region_thresholds] / 5
plt.imshow(region_select)
plt.show()
```

```python
line_image = np.copy(image)
line_image[~thresholds & region_thresholds] = [255,0,0]

plt.imshow(line_image)
plt.show()
```

```python
import cv2
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) #grayscale conversion
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
kernel_size = 3
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
plt.imshow(blur_gray, cmap='gray')
plt.show()
```

```python
low_threshold = 100
high_threshold = 250
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

plt.imshow(edges, cmap='Greys_r')
plt.show()
```

```python
v = np.median(blur_gray)
sigma = 0.33

lower = int(max(0, (1.0 - sigma) * v))
upper = int(min(255, (1.0 + sigma) * v))
edged = cv2.Canny(blur_gray, lower, upper)
plt.imshow(edged, cmap='Greys_r')
plt.show()
```

```python
rho = 1
theta = np.pi/180
threshold = 1
min_line_length = 10
max_line_gap = 1 
line_image = np.copy(image) * 0
```

```python
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
```

```python
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
        
color_edges = np.dstack((edges, edges, edges)) 

combo = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0) 
plt.imshow(line_image)
plt.show()

plt.imshow(combo)
plt.show()
```

```python
mask = np.zeros_like(edges)   
ignore_mask_color = [255, 255, 255]  
imshape = image.shape
vertices = np.array([[(80,imshape[0]),(450, 280), (490, 280), (890,imshape[0])]], dtype=np.int32)
cv2.fillPoly(mask, vertices, ignore_mask_color)
masked_edges = cv2.bitwise_and(edges, mask)

plt.imshow(masked_edges)
plt.show()
```

```python
rho = 2
theta = np.pi/180
threshold = 3
min_line_length = 10
max_line_gap = 8 

line_image2 = np.copy(image) * 0

lines2 = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)

for line in lines2:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image2,(x1,y1),(x2,y2),(255,0,0),10)

# Create a "color" binary image to combine with line image
color_edges2 = np.dstack((edges, edges, edges)) 

# Draw the lines on the edge image
lines_edges2 = cv2.addWeighted(color_edges2, 0.8, line_image2, 1, 0) 
plt.imshow(lines_edges2)
plt.show()
```

```python
final = np.copy(image)
final2 = cv2.addWeighted(final, 0.8, line_image2, 1, 0) 
plt.imshow(final2)
plt.show()
```
